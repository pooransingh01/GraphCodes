```python
#Write a code to find the degree of each vertex, and store it in a dictionary. Sort the keys of the dictionary by the degree stored in the values.
def calculate_vertex_degrees(graph):
    degree_dict = {}
    
    # Iterate through each node in the graph
    for node in graph:
        # The degree is the number of adjacent vertices (length of the adjacency list)
        degree = len(graph[node])
        degree_dict[node] = degree
    
    return degree_dict

   
```

```python
#Given a graph and two vertices, check if they are adjacent. 

def are_adjacent(node1, node2, graph):
    if node1 in graph:
        neighbors = graph[node1]
        for neighbor in neighbors:
            if neighbor == node2:
                return True
    return False
```

```python
#Check if a graph is complete.

def is_complete(graph):
    total_nodes = len(graph)

    for node in graph:
        neighbors = graph[node]

        # In a complete graph, every node is connected to all other nodes
        if len(neighbors) != total_nodes - 1:
            return False

    return True
```

```python
#Given a tree, find the number of leaf nodes.

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class BinaryTree:
    def __init__(self, root):
        self.root = root

    def count_leaf_nodes(self, node):
        if node is None:
            return 0
        if node.left is None and node.right is None:
            print(f"Leaf node found: {node.value}")
            return 1

        leaf_count_left = self.count_leaf_nodes(node.left)
        leaf_count_right = self.count_leaf_nodes(node.right)

        return leaf_count_left + leaf_count_right
```

```python
#Given a tree and a node, find its height.

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left_child = None
        self.right_child = None


class BinaryTree:
    def __init__(self, root_node):
        self.root_node = root_node

    def find_height(self, current_node):
        if current_node is None:
            return "No height"

        left_subtree_height = self.find_height(current_node.left_child)
        right_subtree_height = self.find_height(current_node.right_child)

        current_height = max(left_subtree_height, right_subtree_height) + 1
        print(f"Height at node {current_node.data} is {current_height}")
        return current_height


```

```python
#Given a tree and a node, find its depth.

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left_child = None
        self.right_child = None


class BinaryTree:
    def __init__(self, root_node):
        self.root_node = root_node

    def find_depth(self, current_node, target_value, current_depth=0):
        if current_node is None:
            return "No depth"

        if current_node.data == target_value:
            print(f"Node {current_node.data} has depth {current_depth}")
            return current_depth

        left_result = self.find_depth(current_node.left_child, target_value, current_depth + 1)
        if left_result != -1:
            return left_result

        right_result = self.find_depth(current_node.right_child, target_value, current_depth + 1)
        return right_result


```
